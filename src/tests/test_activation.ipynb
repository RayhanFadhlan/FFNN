# Pengujian Pengaruh Fungsi Aktivasi Yang Berbeda

Pengujian bertujuan untuk menganalisis pengaruh fungsi aktivasi pada FFNN.

```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt


# Add parent directory to path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "../")))

%load_ext autoreload
%autoreload 2

from model import FFNN
from activation import Linear, ReLU, Sigmoid, Tanh, Softmax, LeakyReLU, ELU
from loss import CategoricalCrossEntropy
from utils import load_mnist_data, train_and_evaluate

# Create out dir
os.makedirs("../../out", exist_ok=True)
```

```python
# Load MNIST dataset
# X_train, X_test, y_train, y_test, y_train_onehot, y_test_onehot = load_mnist_data(subset_size=10000)
X_train, X_test, y_train, y_test, y_train_onehot, y_test_onehot = load_mnist_data()


input_size = X_train.shape[1]
output_size = y_train_onehot.shape[1]
hidden_size = 64  # Fixed hidden layer size for all experiments

# Experiment configs
epochs = 10
batch_size = 32
learning_rate = 0.01
```

```python
# Test different activation functions
activations = [
    ("Linear", Linear()),
    ("ReLU", ReLU()),
    ("Sigmoid", Sigmoid()),
    ("Tanh", Tanh()),
    ("LeakyReLU", LeakyReLU(alpha=0.01)),
    ("ELU", ELU(alpha=1.0))
]

activation_histories = []
activation_accuracies = []
trained_models = []

for name, activation_func in activations:
    model_name = f"FFNN with {name} activation"
    print(f"\nTraining model with {name} activation function")
    
    model = FFNN(
        layer_sizes=[input_size, hidden_size, output_size],
        activations=[activation_func, Softmax()],
        loss=CategoricalCrossEntropy()
    )
    
    history, accuracy, _, trained_model = train_and_evaluate(
        model, X_train, X_test, y_train_onehot, y_test, y_test_onehot,
        epochs=epochs, batch_size=batch_size, learning_rate=learning_rate,
        model_name=model_name
    )
    
    activation_histories.append(history)
    activation_accuracies.append(accuracy)
    trained_models.append(trained_model)

# Plot loss curves for all activation functions
plt.figure(figsize=(14, 7))
for i, (name, _) in enumerate(activations):
    plt.plot(activation_histories[i]['loss'], label=f'{name}')
    plt.plot(activation_histories[i]['val_loss'], label=f'validation {name}')

plt.title('Training Loss Comparison of Different Activation Functions')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.savefig('../../out/activation_comparison_loss.png')

# Bar chart of accuracies
plt.figure(figsize=(12, 6))
plt.bar([name for name, _ in activations], activation_accuracies, color='skyblue')
plt.title('Test Accuracy Comparison of Different Activation Functions')
plt.xlabel('Activation Function')
plt.ylabel('Accuracy')
plt.ylim([0.8, 1.0])
plt.xticks(rotation=45)
for i, v in enumerate(activation_accuracies):
    plt.text(i, v + 0.01, f"{v:.4f}", ha='center')
plt.tight_layout()
plt.savefig('../../out/activation_comparison_accuracy.png')

# Plot weight and gradient distributions
for i, (name, _) in enumerate(activations):
    model = trained_models[i]
    
    # Plot weight distributions
    plt.figure(figsize=(12, 5))
    plt.suptitle(f'Weight Distribution with {name} Activation', fontsize=16)
    
    for j, layer in enumerate(model.layers):
        plt.subplot(1, len(model.layers), j+1)
        weights = layer.weights.flatten()
        plt.hist(weights, bins=50, alpha=0.7)
        plt.title(f'Layer {j+1}')
        plt.xlabel('Weight Value')
        plt.ylabel('Frequency')
        
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.savefig(f'../../out/activation_{name}_weight_dist.png')
    
    # Plot gradient distributions
    plt.figure(figsize=(12, 5))
    plt.suptitle(f'Gradient Distribution with {name} Activation', fontsize=16)
    
    for j, layer in enumerate(model.layers):
        plt.subplot(1, len(model.layers), j+1)
        gradients = layer.weights_grad.flatten()
        plt.hist(gradients, bins=50, alpha=0.7)
        plt.title(f'Layer {j+1}')
        plt.xlabel('Gradient Value')
        plt.ylabel('Frequency')
        
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.savefig(f'../../out/activation_{name}_gradient_dist.png')


# Best and worst performing activation functions
best_idx = np.argmax(activation_accuracies)
worst_idx = np.argmin(activation_accuracies)

print(f"\nBest performing: {activations[best_idx][0]} with accuracy {activation_accuracies[best_idx]:.4f}")
print(f"Worst performing: {activations[worst_idx][0]} with accuracy {activation_accuracies[worst_idx]:.4f}")
```
