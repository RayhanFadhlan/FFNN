# Pengujian Pengaruh Learning Rate pada FFNN

Pengujian bertujuan untuk menganalisis pengaruh learning rate pada training FFNN.

```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt

# Add parent directory to path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "../")))

%load_ext autoreload
%autoreload 2

from model import FFNN
from activation import ReLU, Softmax
from loss import CategoricalCrossEntropy
from utils import load_mnist_data, train_and_evaluate

# Create out dir
os.makedirs("../../out", exist_ok=True)
```

```python
# Load MNIST dataset
X_train, X_test, y_train, y_test, y_train_onehot, y_test_onehot = load_mnist_data()

input_size = X_train.shape[1]  # 784 for MNIST
output_size = y_train_onehot.shape[1]  # 10 for MNIST
hidden_size = 64  # Fixed hidden layer size for all experiments

# Experiment configs
epochs = 10
batch_size = 32

# Test different learning rates
learning_rates = [0.001, 0.01, 0.1]
lr_histories = []
lr_accuracies = []
trained_models = []

for lr in learning_rates:
    model_name = f"FFNN with learning rate={lr}"
    print(f"\nTraining model with learning rate={lr}")
    
    model = FFNN(
        layer_sizes=[input_size, hidden_size, output_size],
        activations=[ReLU(), Softmax()],
        loss=CategoricalCrossEntropy()
    )
    
    history, accuracy, _, trained_model = train_and_evaluate(
        model, X_train, X_test, y_train_onehot, y_test, y_test_onehot,
        epochs=epochs, batch_size=batch_size, learning_rate=lr,
        model_name=model_name
    )
    
    lr_histories.append(history)
    lr_accuracies.append(accuracy)
    trained_models.append(trained_model)


# Plot loss curves for all learning rates
plt.figure(figsize=(12, 6))
for i, lr in enumerate(learning_rates):
    plt.plot(lr_histories[i]['loss'], label=f'LR={lr}')

plt.title('Training Loss Comparison of Different Learning Rates')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.savefig('../../out/learning_rate_comparison_loss.png')


# Bar chart of accuracies
plt.figure(figsize=(10, 6))
plt.bar([f'LR={lr}' for lr in learning_rates], lr_accuracies, color='skyblue')
plt.title('Test Accuracy Comparison of Different Learning Rates')
plt.xlabel('Learning Rate')
plt.ylabel('Accuracy')
# plt.ylim([0.8, 1.0])
for i, v in enumerate(lr_accuracies):
    plt.text(i, v + 0.01, f"{v:.4f}", ha='center')
plt.savefig('../../out/learning_rate_comparison_accuracy.png')


for i, lr in enumerate(learning_rates):
    model = trained_models[i]
    
    # Plot weight distributions
    plt.figure(figsize=(12, 5))
    plt.suptitle(f'Weight Distribution with Learning Rate = {lr}', fontsize=16)
    
    for j, layer in enumerate(model.layers):
        plt.subplot(1, len(model.layers), j+1)
        weights = layer.weights.flatten()
        plt.hist(weights, bins=50, alpha=0.7)
        plt.title(f'Layer {j+1}')
        plt.xlabel('Weight Value')
        plt.ylabel('Frequency')
        
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.savefig(f'../../out/lr_{lr}_weight_dist.png')
    
    # Plot gradient distributions
    plt.figure(figsize=(12, 5))
    plt.suptitle(f'Gradient Distribution with Learning Rate = {lr}', fontsize=16)
    
    for j, layer in enumerate(model.layers):
        plt.subplot(1, len(model.layers), j+1)
        gradients = layer.weights_grad.flatten()
        plt.hist(gradients, bins=50, alpha=0.7)
        plt.title(f'Layer {j+1}')
        plt.xlabel('Gradient Value')
        plt.ylabel('Frequency')
        
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.savefig(f'../../out/lr_{lr}_gradient_dist.png')


print("\nACCURACY RESULTS:")
for i, lr in enumerate(learning_rates):
    print(f"Learning Rate = {lr}: {lr_accuracies[i]:.4f}")

# Best and worst learning rates
best_idx = np.argmax(lr_accuracies)
worst_idx = np.argmin(lr_accuracies)

print(f"\nBest performing: Learning Rate = {learning_rates[best_idx]} with accuracy {lr_accuracies[best_idx]:.4f}")
print(f"Worst performing: Learning Rate = {learning_rates[worst_idx]} with accuracy {lr_accuracies[worst_idx]:.4f}")
```
