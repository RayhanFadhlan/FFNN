# Pengujian Pengaruh Inisialisasi Bobot Pada Training FFNN

Pengujian ini bertujuan untuk menganalisis pengaruh inisialisasi nilai bobot pada training FFNN.

```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt

# Add parent directory to path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "../")))

%load_ext autoreload
%autoreload 2

from model import FFNN
from activation import ReLU, Softmax
from loss import CategoricalCrossEntropy
from initializer import ZeroInitializer, UniformInitializer, NormalInitializer, XavierInitializer, HeInitializer
from layer import Layer
from utils import  train_and_evaluate, load_mnist_data

# Create out dir
os.makedirs("../../out", exist_ok=True)
```

```python
def create_model_with_initializer(input_size, hidden_size, output_size, initializer):
    """
    Create a model with the specified initializer for all layers
    """
    model = FFNN(
        layer_sizes=[input_size, hidden_size, output_size],
        activations=[ReLU(), Softmax()],
        loss=CategoricalCrossEntropy()
    )
    
    # Replace the default initializers with the specified one
    model.layers = []
    for i in range(len(model.layer_sizes) - 1):
        model.layers.append(Layer(
            input_size=model.layer_sizes[i],
            output_size=model.layer_sizes[i + 1],
            activation=model.activations[i],
            weight_initializer=initializer
        ))
    
    return model
```

```python
# Load MNIST dataset
# X_train, X_test, y_train, y_test, y_train_onehot, y_test_onehot = load_mnist_data(subset_size=10000)
X_train, X_test, y_train, y_test, y_train_onehot, y_test_onehot = load_mnist_data()


input_size = X_train.shape[1]  # 784 for MNIST
output_size = y_train_onehot.shape[1]  # 10 for MNIST
hidden_size = 64  # Fixed hidden layer size for all experiments

# Experiment config
epochs = 10
batch_size = 32
learning_rate = 0.01

# Testing different weight initialization methods
initializers = [
    ("Zero", ZeroInitializer()),
    ("Uniform", UniformInitializer(low=-0.1, high=0.1, seed=42)),
    ("Normal", NormalInitializer(mean=0.0, std=0.1, seed=42)),
    ("Xavier", XavierInitializer(seed=42)),
    ("He", HeInitializer(seed=42))
]

init_histories = []
init_accuracies = []
trained_models = []

for name, initializer in initializers:
    model_name = f"FFNN with {name} initialization"
    print(f"\nTraining model with {name} initialization")
    
    model = create_model_with_initializer(input_size, hidden_size, output_size, initializer)
    
    history, accuracy, _, trained_model = train_and_evaluate(
        model, X_train, X_test, y_train_onehot, y_test, y_test_onehot,
        epochs=epochs, batch_size=batch_size, learning_rate=learning_rate,
        model_name=model_name
    )
    
    init_histories.append(history)
    init_accuracies.append(accuracy)
    trained_models.append(trained_model)

# Plot loss curves for all initialization methods
plt.figure(figsize=(12, 6))
for i, (name, _) in enumerate(initializers):
    plt.plot(init_histories[i]['loss'], label=f'{name}')

plt.title('Training Loss Comparison of Different Weight Initialization Methods')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.savefig('../../out/init_comparison_loss.png')

# Bar chart of accuracies
plt.figure(figsize=(12, 6))
plt.bar([name for name, _ in initializers], init_accuracies, color='skyblue')
plt.title('Test Accuracy Comparison of Different Weight Initialization Methods')
plt.xlabel('Initialization Method')
plt.ylabel('Accuracy')
plt.ylim([0, 1.0])
plt.xticks(rotation=45)
for i, v in enumerate(init_accuracies):
    plt.text(i, v + 0.01, f"{v:.4f}", ha='center')
plt.tight_layout()
plt.savefig('../../out/init_comparison_accuracy.png')

# Plot weight and gradient distributions for each model
for i, (name, _) in enumerate(initializers):
    model = trained_models[i]
    
    # Initial weight distributions
    plt.figure(figsize=(12, 5))
    plt.suptitle(f'Initial Weight Distribution with {name} Initialization', fontsize=16)
    
    # Recreate a model with the same initializer to show initial weights
    initial_model = create_model_with_initializer(input_size, hidden_size, output_size, initializers[i][1])
    
    for j, layer in enumerate(initial_model.layers):
        plt.subplot(1, len(initial_model.layers), j+1)
        weights = layer.weights.flatten()
        plt.hist(weights, bins=50, alpha=0.7)
        plt.title(f'Layer {j+1}')
        plt.xlabel('Weight Value')
        plt.ylabel('Frequency')
        
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.savefig(f'../../out/init_{name}_initial_weight_dist.png')
    
    # Final weight distributions
    plt.figure(figsize=(12, 5))
    plt.suptitle(f'Final Weight Distribution with {name} Initialization', fontsize=16)
    
    for j, layer in enumerate(model.layers):
        plt.subplot(1, len(model.layers), j+1)
        weights = layer.weights.flatten()
        plt.hist(weights, bins=50, alpha=0.7)
        plt.title(f'Layer {j+1}')
        plt.xlabel('Weight Value')
        plt.ylabel('Frequency')
        
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.savefig(f'../../out/init_{name}_final_weight_dist.png')
    
    # Gradient distributions
    plt.figure(figsize=(12, 5))
    plt.suptitle(f'Gradient Distribution with {name} Initialization', fontsize=16)
    
    for j, layer in enumerate(model.layers):
        plt.subplot(1, len(model.layers), j+1)
        gradients = layer.weights_grad.flatten()
        plt.hist(gradients, bins=50, alpha=0.7)
        plt.title(f'Layer {j+1}')
        plt.xlabel('Gradient Value')
        plt.ylabel('Frequency')
        
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.savefig(f'../../out/init_{name}_gradient_dist.png')
```

```python
print("\nACCURACY RESULTS:")
for i, (name, _) in enumerate(initializers):
    print(f"{name} Initialization: {init_accuracies[i]:.4f}")

# Best and worst performing initialization methods
best_idx = np.argmax(init_accuracies)
worst_idx = np.argmin(init_accuracies)

print(f"\nBest performing: {initializers[best_idx][0]} with accuracy {init_accuracies[best_idx]:.4f}")
print(f"Worst performing: {initializers[worst_idx][0]} with accuracy {init_accuracies[worst_idx]:.4f}")
```
