# Pengujian Pengaruh Depth dan Width pada FFNN

Pengujian betujuan untuk menganalisis pengaruh kedalaman jaringan (depth) 
dan lebar jaringan (width) terhadap performa model FFNN.

```python
import os
import sys

import matplotlib.pyplot as plt

# Add parent directory to path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "../")))

%load_ext autoreload
%autoreload 2

from activation import ReLU, Softmax
from loss import CategoricalCrossEntropy
from model import FFNN
from utils import load_mnist_data, train_and_evaluate

# Create out dir
os.makedirs("../../out", exist_ok=True)
```

```python
# Load MNIST dataset
# X_train, X_test, y_train, y_test, y_train_onehot, y_test_onehot = load_mnist_data(subset_size=10000)
X_train, X_test, y_train, y_test, y_train_onehot, y_test_onehot = load_mnist_data()

input_size = X_train.shape[1]  # 784 for MNIST
output_size = y_train_onehot.shape[1]  # 10 for MNIST

# Experiment configs
epochs = 10
batch_size = 32
learning_rate = 0.01
```

## Experiment 1: Varying Width

```python
# Test 3 different widths with fixed depth
widths = [32, 64, 128]
width_histories = []
width_accuracies = []

for width in widths:
    model_name = f"FFNN (1 hidden layer, {width} neurons)"
    print(f"\nTraining model with width={width}")

    model = FFNN(
        layer_sizes=[input_size, width, output_size],
        activations=[ReLU(), Softmax()],
        loss=CategoricalCrossEntropy(),
    )

    history, accuracy, _, _ = train_and_evaluate(
        model,
        X_train,
        X_test,
        y_train_onehot,
        y_test,
        y_test_onehot,
        epochs=epochs,
        batch_size=batch_size,
        learning_rate=learning_rate,
        model_name=model_name,
    )

    width_histories.append(history)
    width_accuracies.append(accuracy)

# Plot width experiment results
plt.figure(figsize=(12, 6))
for i, width in enumerate(widths):
    plt.plot(width_histories[i]["loss"], label=f"Width={width}")

plt.title("Training Loss vs. Width (1 Hidden Layer)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.savefig("../../out/width_comparison_loss.png")

# Bar chart of accuracies
plt.figure(figsize=(10, 6))
plt.bar([str(w) for w in widths], width_accuracies, color="skyblue")
plt.title("Test Accuracy vs. Width (1 Hidden Layer)")
plt.xlabel("Width (neurons)")
plt.ylabel("Accuracy")
plt.ylim([0.9 * min(width_accuracies), 1.0])
for i, v in enumerate(width_accuracies):
    plt.text(i, v + 0.01, f"{v:.4f}", ha="center")
plt.savefig("../../out/width_comparison_accuracy.png")
```

## Experiment 2: Varying Depth

```python
# Test 3 different depths with fixed width
depths = [1, 2, 3]  # Number of hidden layers
depth_histories = []
depth_accuracies = []

for depth in depths:
    layer_sizes = [input_size] + [64] * depth + [output_size]
    activations = [ReLU()] * depth + [Softmax()]

    model_name = f"FFNN ({depth} hidden layers, 64 neurons each)"
    print(f"\nTraining model with depth={depth}")

    model = FFNN(
        layer_sizes=layer_sizes,
        activations=activations,
        loss=CategoricalCrossEntropy(),
    )

    history, accuracy, _, _ = train_and_evaluate(
        model,
        X_train,
        X_test,
        y_train_onehot,
        y_test,
        y_test_onehot,
        epochs=epochs,
        batch_size=batch_size,
        learning_rate=learning_rate,
        model_name=model_name,
    )

    depth_histories.append(history)
    depth_accuracies.append(accuracy)

# Plot depth experiment results
plt.figure(figsize=(12, 6))
for i, depth in enumerate(depths):
    plt.plot(depth_histories[i]["loss"], label=f"Depth={depth}")

plt.title("Training Loss vs. Depth (Width=64)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.savefig("../../out/depth_comparison_loss.png")

# Bar chart of accuracies
plt.figure(figsize=(10, 6))
plt.bar([str(d) for d in depths], depth_accuracies, color="lightgreen")
plt.title("Test Accuracy vs. Depth (Width=64)")
plt.xlabel("Depth (hidden layers)")
plt.ylabel("Accuracy")
plt.ylim([0.9 * min(depth_accuracies), 1.0])
for i, v in enumerate(depth_accuracies):
    plt.text(i, v + 0.01, f"{v:.4f}", ha="center")
plt.savefig("../../out/depth_comparison_accuracy.png")
```

## Comparison Analysis

```python
print("\nWIDTH EXPERIMENT RESULTS:")
for i, width in enumerate(widths):
    print(f"Width={width}: Accuracy={width_accuracies[i]:.4f}")

print("\nDEPTH EXPERIMENT RESULTS:")
for i, depth in enumerate(depths):
    print(f"Depth={depth}: Accuracy={depth_accuracies[i]:.4f}")

# Plot combined analysis
plt.figure(figsize=(14, 7))

# Plot width comparison
plt.subplot(1, 2, 1)
bars1 = plt.bar([str(w) for w in widths], width_accuracies, color="skyblue")
plt.title("Accuracy vs. Width (1 Hidden Layer)")
plt.xlabel("Width (neurons)")
plt.ylabel("Accuracy")
plt.ylim([0.9 * min(width_accuracies), 1.0])
for bar in bars1:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2.0,
        height + 0.005,
        f"{height:.4f}",
        ha="center",
        va="bottom",
    )

# Plot depth comparison
plt.subplot(1, 2, 2)
bars2 = plt.bar([str(d) for d in depths], depth_accuracies, color="lightgreen")
plt.title("Accuracy vs. Depth (Width=64)")
plt.xlabel("Depth (hidden layers)")
plt.ylabel("Accuracy")
plt.ylim([0.9 * min(depth_accuracies), 1.0])
for bar in bars2:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2.0,
        height + 0.005,
        f"{height:.4f}",
        ha="center",
        va="bottom",
    )

plt.tight_layout()
plt.savefig("../../out/depth_width_comparison.png")
```
